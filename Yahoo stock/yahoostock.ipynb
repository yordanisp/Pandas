### Yahoo Data Using Pandas ###

```python
# import libraries
import pandas as pd
import pandas_datareader as pdr
import sys
%matplotlib inline
```

```python
print('Python: ' + sys.version.split('|')[0])
print('Pandas: ' + pd.__version__)
print('Pandas_datareader: ' + pdr.__version__)
```

```python
# Pull Yahoo Stock Data
# Select some dates to play with
start = pd.to_datetime('2020-07-01')
end = pd.to_datetime('2020-07-31')

# Pick your favorite stock!
df = pdr.get_data_yahoo('M', start, end)
df

df.info
```

```python
# Open vs Close prices
df[['Open', 'Close']].plot(figsize=(15,5));
```

```python
# Slice Your Data by a Date Range
df.loc['2020-03-15':'2020-03-24', ['Open', 'Close']].plot(figsize=(15,5));
```

```python
# Chart the Percentage change of Closing prices
df.Close.pct_change().plot(figsize=(15,5));
```

```python
# Get the median percentage change of Closing prices
df.Close.pct_change().median()
```
